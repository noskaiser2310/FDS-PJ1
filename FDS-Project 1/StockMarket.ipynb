```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Stock Market Dataset.csv",index_col=0)
df = df.sort_index(ascending=False)
df.head(10)
```

```python
df.info()
```

```python
columns = df.columns

columns
```

```python
df.describe(include='O')
```

```python
df_o = ['Bitcoin_Price','Ethereum_Price','S&P_500_Price','Nasdaq_100_Price','Gold_Price','Platinum_Price','Berkshire_Price']
```

```python
df[df_o] = df[df_o].replace(',', '', regex=True).astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
insull_column = df.isnull().sum() / len(df) * 100
insull_column.sort_values(ascending=False)
```

```python
df.bfill(inplace=True)
```

```python
df.describe()
```

```python
df['Platinum_Vol.'] = df['Platinum_Vol.'].fillna(0)
```

```python
columns = [col for col in columns if col != 'Date']
for col in columns :
    df[col] = df[col].apply(lambda x: 0 if x < 0 else x)
```

```python
df.to_csv('Stock Market Clean.csv')
```
